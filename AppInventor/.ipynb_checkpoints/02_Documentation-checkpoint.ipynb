# Built-in Blocks

## Control Blocks 

1. if & else if 



Test a given condition. If the result is true, performs the actiosn in the -then sequence of blocks; otherwise tests the statement in the -else if section. If the result is true, performs the action in the -then sequence of blocks; otherwise, performs the actions in the -else sequence of blocks. 

2. for each number from to 

Runs the block in the do section for each numeric value in the range starting from *from* and ending at *to*, incrementaing **number** by the value of by each time. Use the given variable name, **number**, to refer to the current value. You can change the  name **number** to something else if you wish. 

3. for each item in list 

Runs the blocks in the do section for each item in the list. Use the given variable name, **item**, to refer to the current list item. You can change the name **item** to something else if you wish. 

4. for each key with value in dictionary 

Runs the block in the do section for each key-value entry in the dictionary. Use the given variables, **key** and **value**, to refer to the key and value of the current dictionary entry. You can change the names **key** and **value** to something else if you wish. 

5. while 

Tests the -test condition. if true, performs the action given in -do, then tests agian. When test is false, the block ends and the action given in -do is no longer performed. 

6. if then else 

Tests a given condition. If the statement is true, performs the actions in the then-return sequence of blocks and returns teh then-return value; otherwise, performs the action in the else-return sequence of blocks and return the else-return value. This block is similar to the ternary to the ternary operator. 

7. do with result 

Sometimes in a procedure or another block of code, you may need to do something and return something, but for various reason you may choose to use this block instead of creating a new procedure. 

8. evaluate but ignore result 

Provides a "dummy socket" for fitting a block that has a plug on its left into a place where there is no socket, such as one of the sequence of blocks in the do part of a procedufre of an if block. The block you fit in will be run, but its returned result will be ignored. 

9. open another screen 

Opens the screen with the provided name. 

The screenName must be one of the Screens created using the Designer. The screenName should be selected from the connected screen name dropdown block. If you do open another screen, you should close it when returning to your main screen to free system memory. Failure to close a screen upon leaving it will eventually lead to memory problems. 

App developers should never close Screen1 or use this block to return to Screen1. Use the close screen block instead. 

10. open another screen with start value 

Opens another screen and passes a value to it.

11. get plain start text 

Return the plain text that was passed to this screen when it was started by another app. If no value was passed, it returns the empty text. For multiple screen apps, use get start value rather than get plain start text.

12. get start value 

Returns the start value given to the current screen. 

This value is given from using **open anohter scrren with start value** or **close screen with value**.

13. clsoe screen

Closes the current screen. 

14. close screen with plain text 

Closes the current screen and passes text to the app that opened this one. This command is for returning text to non-app Inventor activities, not to App Inventor screens. 

15. close screen with values 

Closes the current screen and return a value to the screen that opened this one. 

16. close application

Closes the application. 

17. break 

When looping using the for range, for each, or while blocks it sometimes useful to be able to exit the loop early. The break allows you to escape the loop. When executed, this exit the loop and continue the app with statements that occur after the loop in the blocks. 

## Logic Blocks 

1. true 

Represents the constant value true. Use it for setting boolean property values of components, or as the value of variable that represent a condition. 

2. false 

Represents the constant value flase. Use it for setting boolean property values of components, or as the value of variable that represent a condition. 

3. not 

Performs logical negation, returning false if the input is true, and true if the input is false. 

4. ={#=}

Test whether its arguments are equal.

- Two numbers are equal if they are numerically eqaul, for example 1 is equal to 1. 0.
- Two text blocks are equal if they have the same characters in the same order, with the same case. For example, banana is not equal to Banana.
- Numbers and text are equal if the number is numerically equal to a number that would be printed with that text. For example, 12.0 is equal to the result of joining the first character of 1A to the last character of Teafor2.
- Two lists are equal if they have the same number of elements and the corresponding elements are equal.

5. != {#not=}

Tests to see whether two arguments are not equal. 

6. and 

Tests whether all of a set of logical conditions are true. The result is true if and only if all the tested conditions are true. The number of tests can be expanded using the mutator. The conditions are tested left to right, and the testing stops as soon as one of the conditions is false. If there are no conditions to test, then the result is true. You can consider this to be a logician’s joke.

6. or 

Tests whether any of a set of logical conditions are true. The result is true if one or more of the tested conditions are true. The number of tests can be expanded using the mutator. The conditions are tested left to right, and the testing stops as soon as one of the conditions is true. If there are no conditions to test, then the result is false.

## Math Blocks

1. Basic Number Block 

Can be used as any positive or negative number. Clicking on the "0" in the block will allow you to change the number. The block supports normal base-10 numbers as well as C-like prefixes for other number bases.

2. Radix Number Block 

Represents a base-10 number. Clicking on the "10" will allow you to change the number. Clicking the dropdown will allow you to input a number in a different number base. The number will then be "translated" into decimal. 

3. ={#=}

Tests whether two numbers are equal and returns true or false.

4. !={#not=}

Tests whether two numbers are not equal and returns true or false.

5. \> 

Tests whether the first number is greater than the second number and returns true or false.

6. \>= 

Tests whether the first number is greater than or equal to the second number and returns true or false.

7. <

Tests whether the first number is less than the second number and returns true or false.

8. <= 

Tests whether the first number is less than or equal to the second number and returns true or false.

9. + 

Returns the result of adding any amount of blocks that have a number value together. Blocks with a number value include the basic number block, length of list or text, variables with a number value, etc. This block is a mutator and can be expanded to allow more numbers in the sum.

10. -

11. X 

Returns the result of multiplying any amount of blocks that have a number value together. It is a mutator block and can be expanded to allow more numbers in the product.

12. / 

Returns the result of dividing the first number by the second.

13. ^ 

Returns the result of the first number raised to the power of the second.

14. random integer 

Returns a random integer value between the given values, inclusive. The order of the arguments doesn’t matter.

15. random fraction 

Returns a random value between 0 and 1.

16. Random set seed to 

Use this block to generate repeatable sequences of random numbers. You can generate the same sequence of random numbers by first calling random set seed with the same value. This is useful for testing programs that involve random values.

17. min

Returns the smallest value of a set of numbers. If there are unplugged sockets in the block, min will also consider 0 in its set of numbers. This block is a mutator and a dropdown.

18. max

Returns the largest value of a set of numbers. If there are unplugged sockets in the block, max will also consider 0 in its set of numbers. This block is a mutator and a dropdown.

19. square root 

Returns the square root of the given number.

20. absolute

Returns the absolute value of the given number.

21. neg 

Returns the negative of a given number.

22. log

Returns the natural logarithm of a given number, that is, the logarithm to the base e (2.71828…).

23. e^

Returns e (2.71828…) raised to the power of the given number.

24. round

Returns the given number rounded to the closest integer. If the fractional part is < .5 it will be rounded down. It it is > .5 it will be rounded up. If it is exactly equal to .5, numbers with an even whole part will be rounded down, and numbers with an odd whole part will be rounded up. (This method is called round to even.)

25. ceiling

Returns the smallest integer that’s greater than or equal to the given number.

26. floor

Returns the greatest integer that’s less than or equal to the given number.

27. modulo

Modulo(a,b) is the same as remainder(a,b) when a and b are positive. More generally, modulo(a,b) is defined for any a and b so that (floor(a/b)× b) + modulo(a,b) = a. For example, modulo(11, 5) = 1, modulo(-11, 5) = 4, modulo(11, -5) = -4, modulo(-11, -5) = -1. Modulo(a,b) always has the same sign as b, while remainder(a,b) always has the same sign as a.

28. remainder

Remainder(a,b) returns the result of dividing a by b and taking the remainder. The remainder is the fractional part of the result multiplied by b.

For example, remainder(11,5) = 1 because

11 / 5 = 2 1⁄5

In this case, 1⁄5 is the fractional part. We multiply this by b, in this case 5 and we get 1, our remainder.

Other examples are remainder(-11, 5) = -1, remainder(11, -5) = 1, and remainder(-11, -5) = -1.

29. quotient

Returns the result of dividing the first number by the second and discarding any fractional part of the result.

30. sin

Returns the sine of the given number in degrees.

31. cos

Returns the cosine of the given number in degrees.

32. tan

Returns the tangent of the given number in degrees.

33. asin

Returns the arcsine of the given number in degrees.

34. acos

Returns the arccosine of the given number in degrees.

35. atan

Returns the arctangent of the given number in degrees.

36. atan2

Returns the arctangent of y/x, given y and x.

37. convert radians to degrees

Returns the value in degrees of the given number in radians. The result will be an angle in the range [0, 360)
convert degrees to radians

Returns the value in radians of the given number in degrees. The result will be an angle in the range [-π , +π)
format as decimal

Formats a number as a decimal with a given number of places after the decimal point. The number of places must be a non-negative integer. The result is produced by rounding the number (if there were too many places) or by adding zeros on the right (if there were too few).

38. is a number?

Returns true if the given object is a number, and false otherwise.
convert number

Takes a text string that represents a positive integer in one base and returns a string that represents the same number is another base. For example, if the input string is 10, then converting from base 10 to binary will produce the string 1010; while if the input string is the same 10, then converting from binary to base 10 will produce the string 2. If the input string is the same 10, then converting from base 10 to hex will produce the string A.

39. Bitwise And

Takes two numbers and compares each pair of bits. Each bit of the result is 1 only if the corresponding bits of both operands are 1.

Example:
  	Decimal 	Binary (internal representation)
  	6 	0 1 1 0
  	3 	0 0 1 1
Result: 	2 	0 0 1 0
Bitwise Or (Inclusive)

Takes two numbers and compares each pair of bits. Each bit of the result is 1 if either of the corresponding bits in each operand is 1.

Example:
  	Decimal 	Binary (internal representation)
  	6 	0 1 1 0
  	3 	0 0 1 1
Result: 	7 	0 1 1 1
bitwise Or (Exclusive)

Takes two numbers and compares each pair of bits. Each bit of the result is 1 only if one corresponding bit in the operands is 1 and the other is 0.
