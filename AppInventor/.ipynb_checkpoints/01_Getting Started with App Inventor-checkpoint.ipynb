# Setup computer and phone/tablet or emulator for live testing as you build 

## iPhone or Android phone (or tablet) and a Wi-Fi internet connection.

1. Step 1 : Download and install the MIT App Inventor Companion app on your Android or iOS device.
2. Step 2 : Connect both your computer and your device to the **same** Wi-Fi network. 
3. Step 3 : Open an App Inventor project and connect it to your device.  

Go to App Inventor and open a project. Then choose "Connect" and "AI Companion" From the top menu in your browser. A dialog with a QR code will appear on your PC screen. On your device, launch the MIT App Compnion app just as you would do any app. Then click the "Scan QR code" button on the Companion, and scan the code in the App Inventor window. 

## Emulator 

If you do not have an Anroid phone or tablet, you can still build apps with App Inventor. App Inventor provides an Android emulator, which just like an Android but appears on your computer screen. So you can test your apps on an emulator and still dstribute the app to others. 

1. Step 1 : Install the App Inventor Setup Software 
2. Step 2 : Launch aiStarter with command  "/usr/google/appinventor/commands-for-Appinventor/aiStarter &"
3. Step 3 : Open an App Inventor project and connect it to the emulator

Go to App Inventor and open a project. Then from App Inventor's menu, go to the Connect Menu and click the Emulator option. 

# Designer and Blocks Editor

## Designer 

The Designer lets you create the app's interface : 

- Palette : Find your components and drag them to the Viewer to add them to your app. 
- Designer Button : Click from any tab to go to the Designer tab. 
- Properties : Select a Component in the Components List to change its properties (color, size, behavior) here. 
- Viewer : Drag components from the Palette to the Viewer to see what tour app will look like. 

## Blocks Editor 

The Blocks Editor lets you program the app's behavior by putting blocks together : 

- Built-in Drawers : Find Blocks for gerneral behaviors you may want to add to your app and drag them to the Blocks Viewer. 
- Blocks Button : Cick from any tab to go to the Blcoks tab. 
- Component-Specific Drawer : Find Blocks for behaviors for specific Components and drag them to the Block Viewer.
- Viewer : Drag Blocks from the Drawer to the Blocks Viewer to build relationships and behavior.
